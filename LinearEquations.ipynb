```python
import matplotlib.pyplot as plt
from matplotlib.ticker import MultipleLocator
import numpy as np
```

```python
# Two example plots
fig = plt.figure()
ax1 = fig.add_subplot(1,1,1)
ax2 = fig.add_subplot(1,1,1)

spacing = 1 # This can be your user specified spacing. 
minorLocator = MultipleLocator(spacing)
x = [0, 1100]
y = [550, 0]
ax1.plot(x, y)
# Set minor tick locations.
ax1.yaxis.set_minor_locator(minorLocator)
ax1.xaxis.set_minor_locator(minorLocator)
# Set grid to use minor tick locations. 
ax1.grid(which = 'minor')

spacing = 1
minorLocator = MultipleLocator(spacing)
x = [0, 100, 450, 900]
y = [900, 800, 450, 0]
ax2.plot(x, y)
# Set minor tick locations.
ax2.yaxis.set_minor_locator(minorLocator)
ax2.xaxis.set_minor_locator(minorLocator)
# Set grid to use minor tick locations. 
ax2.grid(which = 'minor')

plt.show()
```

```python
# Coordinates (X, Y)
x = np.arange(-4, 4)
y = np.arange(-4, 4)

# print(x, y)

# Plot on graph
plt.plot(y, x, color='red', linestyle='solid')
plt.plot(y, -x, color='blue', linestyle='solid')
plt.plot(y, 5*x, color='green', linestyle='solid')
plt.plot(y, -5*x, color='orange', linestyle='solid')
plt.plot(y, 10*x, color='purple', linestyle='solid')
plt.plot(y, -10*x, color='cyan', linestyle='solid')

# Add a title
plt.title('Plotting Linear Equations')

# Add x and y label
plt.xlabel('x axis')
plt.ylabel('y axis') 

# Show grid
plt.grid()
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
x = np.linspace(-5,5,100)
ax.spines['left'].set_position('center')
ax.spines['bottom'].set_position('center')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')



plt.plot(x, 2*x+1, '-r', label='y=2x+1')
plt.plot(x, 2*x-1,'-.g', label='y=2x-1')
plt.plot(x, 2*x+3,':b', label='y=2x+3')
plt.plot(x, 2*x-3,'--m', label='y=2x-3')
plt.legend(loc='upper left')
plt.grid(which='major', axis='both', linestyle='-')
plt.gca().set_aspect("equal")
plt.gcf().set_size_inches(12, 20)
plt.show()
```

```python
# 100 linearly spaced numbers
x = np.linspace(-5,5,100)

# the function, which is y = x^2 here
y = x**2

# setting the axes at the centre
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.spines['left'].set_position('center')
ax.spines['bottom'].set_position('zero')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')

# plot the function
plt.plot(x,y, 'r')

# show the plot
plt.show()
```

```python
# Funcin cuadrtica.
def f1(x):
    return 2*(x**2) + 5*x - 2
# Funcin lineal.
def f2(x):
    return 4*x + 1
# Valores del eje X que toma el grfico.
x = range(-10, 15)
# Graficar ambas funciones.
plt.plot(x, [f1(i) for i in x])
plt.plot(x, [f2(i) for i in x])
# Establecer el color de los ejes.
plt.axhline(0, color="black")
plt.axvline(0, color="black")
# Limitar los valores de los ejes.
plt.xlim(-10, 10)
plt.ylim(-10, 10)
# Guardar grfico como imgen PNG.
plt.savefig("output.png")
# Mostrarlo.
plt.grid()
plt.show()
```
